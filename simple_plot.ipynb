# Simple Plot

```python
#loading required packages
import matplotlib.pyplot as plt
from numpy import pi, sin, cos, linspace
from matplotlib.ticker import AutoMinorLocator

#Turn on the interactive features of matplotlib
#%matplotlib widget
```

```python
#generate data to plot
theta = linspace(0,1.5*pi,100)
ys = sin(theta)
yc = cos(theta)
```

```python
#plotting the data
fig, ax = plt.subplots()

plt.plot(theta,ys,'--', label=r'$\sin(\theta)$')
plt.plot(theta,yc, label=r'$\cos(\theta)$')
plt.xlabel(r'$\theta$ (radian)')
plt.ylabel(r'$f(\theta)$')
plt.title('Simple Plot')

ax.xaxis.set_minor_locator(AutoMinorLocator())

plt.legend() #loc="lower left"

#saving the plot
fig.savefig("simple_plot.png",format='png')
```

# Multiple plot

```python
fig,ax = plt.subplots(1,2,figsize=(10,4))

ax[0].plot(theta,ys)
ax[0].set_xlabel(r'$\theta$(radian)')
ax[0].set_ylabel(r'$\sin(\theta)$')

ax[1].plot(theta,yc) 
ax[1].set_xlabel(r'$\theta$(radian)')
ax[1].set_ylabel(r'$\cos(\theta)$')

plt.tight_layout()
```
